```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\tejas\Downloads\tourism_data.csv')
```

```python
df.head()
```

Identifying Missing Values

```python
print('Missing values before cleaning the data: ')
print(df.isnull().sum())
```

Now lets deal with those missing values

```python
numeric_col = ['Count' , 'GDP' , 'Population']
df[numeric_col] = df[numeric_col].fillna(df[numeric_col].mean())
```

```python
df['CountryJoin'].fillna(df['CountryJoin'].mode()[0],inplace=True)
```

```python
print('Missing values after cleaning the data : ')
print(df.isnull().sum())
```

# Exploratory Data Analysis (EDA)

```python
# Distribution of tourist arrivals across countries
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Country' , order=df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Tourist Arrivals by Country') 
plt.xlabel('Country')
plt.ylabel('Number of tourist Aarrival')
plt.show()
```

```python
# Distribution of tourist arrivals across months
plt.figure(figsize=(10,5))
sns.countplot(data=df , x='Month' , order=df['Month'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Months')
plt.ylabel('Tourist Arrival')
plt.show()
```

```python
df.columns
```

```python
# Descriptive Statistics
print("\nDescriptive Statistics of Numeric Columns:")
print(df.describe())
```

```python
# Identify Popular Destinations
popular_destinations = df.groupby('Country')['Count'].sum().sort_values(ascending=False)
print("\nTop 5 Popular Tourist Destinations:")
print(popular_destinations.head())
```

```python
# Seasonal Trends
plt.figure(figsize=(10,6))
sns.lineplot(data=df , x='Seasons' , y='Count')
plt.title('Seasons Trend in Tourism')
plt.show()
```

```python
# Comparative Analysis
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Count', hue='Country', data=df)
plt.title('Comparative Analysis of Tourist Arrivals by Country')
plt.xlabel('Year')
plt.ylabel('Number of Tourist Arrivals')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 8))
total_counts_by_country = df.groupby('Country')['Count'].sum().sort_values(ascending=False)
sns.barplot(x=total_counts_by_country.values, y=total_counts_by_country.index, palette='viridis')
plt.title('Total Tourist Counts by Country')
plt.xlabel('Total Tourist Count')
plt.ylabel('Country')
plt.show()
```

Some Insights from the data

1)Top 5 Popular Tourist Destinations : Countries like Argentina, United States, Germany , Italy and Uruguay  are among the top destinations for tourists. Argentina and US seem to attract a higher number of tourists compared to other countries.

2)Seasonal Trends : Seasonal Trends: During winter, there is a higher number of tourists compared to other seasons. There is a decline in tourist counts during summer and fall, and from spring onward, we observe an upward trend in tourist numbers. .

3)Tourist Counts Over Years : While we have data for multiple years, it's essential to analyze the trend of tourist counts over time to identify any significant growth or decline in tourism for specific countries. We observe that the total number of tourist counts over the year is higher in Austria, Denmark, Switzerland, Norway, and Russia.
